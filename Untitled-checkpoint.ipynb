# Error Functions

```python
def L1(schemes,N):
    error_matrix = []
    for i in (schemes[1:]):
        error = 1.0/N*np.sum(abs(schemes[0] - i))
        error_matrix.append(error)
    return error_matrix

def L2(schemes,N):
    error_matrix = []
    for i in (schemes[1:]):
        error = 1.0/N*np.sqrt(np.sum((schemes[0]-i)**2))
        error_matrix.append(error)
    return error_matrix

def Linf(schemes):
    error_matrix = []
    for i in (schemes[1:]):
        error = max(abs(schemes[0]-i))
        error_matrix.append(error)
    return error_matrix

def Errors(anal,num):
    # err_1 = L1(anal,num)
    err_2 = L2(anal,num)
    err_inf = Linf(anal,num)
    print('Error based on L1: ', err_1, ', Error based on L2: ', err_2, ', Error base on L_inf: ', err_inf)
```

```python
# Error functions

def ERROR_L1(num0,num_end,N):
    err_init = L1(num0,N)
    err_final = L1(num_end,N)
    res = np.concatenate([err_init,err_final])
    return res
def ERROR_L2(num0,num_end,N):
    err_init = L2(num0,N)
    err_final = L2(num_end,N)
    res = np.concatenate([err_init,err_final])
    return res
def ERROR_inf(num0,num_end):
    err_init = Linf(num0)
    err_final = Linf(num_end)
    res = np.concatenate([err_init,err_final])
    return res

def All_errors(num0,num_end,N):
    a1 = ERROR_L1(num0,num_end,N)
    a2 = ERROR_L2(num0,num_end,N)
    a3 = ERROR_inf(num0,num_end)
    print(a1)
    print(a2)
    print(a3)
    return a1,a2,a3

grid100_1_09,grid100_2_09,grid100_3_09 = All_errors(num100_09_1,num100_09_10,100)
print()
grid100_1_05,grid100_2_05,grid100_3_05 = All_errors(num100_05_1,num100_05_10,100)
print()
grid100_1_005,grid100_2_005,grid100_3_005 = All_errors(num100_005_1,num100_005_10,100)
```

```python
time_error = np.array([1.0,100.0])
def plot_L(coarse,medium,fine,time,index,CFL,Lax=True):
    plt.plot(time_error,(coarse[0],coarse[3]),label='grid 100, Upwind')
    plt.plot(time_error,(medium[0],medium[3]),label='grid 400')
    plt.plot(time_error,(fine[0],fine[3]),label='grid 1600')
    plt.plot(time_error,(coarse[1],coarse[4]),label='grid 100 Lax-Wendroff')
    plt.plot(time_error,(medium[1],medium[4]),label='grid 400 Lax-Wendroff')
    plt.plot(time_error,(fine[1],fine[4]),label='grid 1600 Lax-Wendroff')
    if Lax == True :
        plt.plot(time_error,(coarse[2],coarse[5]),label='grid 100 Lax-Friedrichs')
    plt.plot(time_error,(medium[2],medium[5]),label='grid 400 Lax-Friedrichs')
    plt.plot(time_error,(fine[2],fine[5]),label='grid 1600 Lax-Friedrichs')
    plt.grid(True, linestyle='--', color='gray', alpha=0.7)
    plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
    plt.title(f'CFL={CFL}, L{index}')
    plt.xlabel('Time')
    plt.ylabel('Error')
    plt.legend(fontsize=7)
```
